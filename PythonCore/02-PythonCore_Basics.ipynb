# تعریف کلی برنامه‌نویسی  

**برنامه‌نویسی** فرآیند **نوشتن دستورات** برای کامپیوتر است تا بتواند **وظایف مشخصی** را اجرا کند. برنامه‌نویسی شامل **طراحی الگوریتم‌ها**، **کدنویسی** و **اشکال‌زدایی** است تا خروجی مطلوب حاصل شود.  

## ارتباط برنامه‌نویسی با الگوریتم‌ها  

یک **الگوریتم** مجموعه‌ای **مرحله‌به‌مرحله** از دستورالعمل‌ها است که به کامپیوتر می‌گوید چگونه یک مسئله را حل کند. هنگام برنامه‌نویسی، باید این مراحل را **به‌صورت دقیق** مشخص کنیم تا سیستم بتواند **بدون خطا** آن‌ها را اجرا کند.  

### نکات مهم در برنامه‌نویسی  
- **کامپیوتر خودکار عمل نمی‌کند**؛ باید **جزء به جزء** به آن بگوییم که چه کاری انجام دهد.  
- **الگوریتم‌ها پایه برنامه‌نویسی هستند** و بدون الگوریتم، برنامه‌نویسی ناممکن است.  
- **هر دستور باید مشخص و دقیق باشد**؛ مبهم بودن کد باعث خطا می‌شود.  
- **بهینه‌سازی الگوریتم** باعث افزایش سرعت و کارایی برنامه می‌شود.

# متغیرها و عبارات در پایتون  

## **متغیر (Variable) چیست؟**  
**متغیر** در پایتون یک **نام** است که برای ذخیره داده‌ها استفاده می‌شود. هر متغیر می‌تواند **عدد، متن، لیست، دیکشنری و انواع دیگر داده‌ها** را نگه دارد.  

## **اصول نام‌گذاری متغیرها**  
- **استفاده از نام‌های معنی‌دار:** نام متغیرها باید مرتبط با داده یا عملکردشان باشد.  
- **رعایت قوانین نام‌گذاری:**  
  - فقط شامل **حروف لاتین، اعداد، و زیرخط (`_`)** باشد.  
  - نباید با **عدد شروع شود**.  
  - **حساس به حروف بزرگ و کوچک** است (`score` و `Score` دو مقدار متفاوت هستند).  
- **سبک‌های استاندارد نام‌گذاری:**  
  - **Snake Case:** `user_name`, `total_price`  
  - **Camel Case:** `userName`, `totalPrice`  
- **اجتناب از استفاده از نام‌های رزرو شده:** نباید از کلمات کلیدی پایتون (`class`, `def`, `return`) به عنوان نام متغیر استفاده شود.  
- **استفاده از `_` برای متغیرهای خصوصی:** `_private_variable` فقط برای استفاده داخلی در کلاس‌ها.  

# **عبارات (Expression) در پایتون**  
یک **Expression** در پایتون ترکیبی از **متغیر‌ها، عملگرها، توابع و مقادیر** است که یک مقدار نهایی تولید می‌کند.  

## **ویژگی‌های عبارات در پایتون**  
- **حساس به حروف است:** `score` و `Score` دو مقدار کاملاً متفاوت هستند.  
- **امکان ترکیب عملیات ریاضی و منطقی:** می‌توان چندین محاسبه را در یک عبارت ترکیب کرد.  
- **امکان استفاده از کدهای تو در تو:** عبارات می‌توانند شامل **توابع و عملیات چندگانه** باشند.  

## **کدهای تو در تو (Nested Expressions)**  
کدهای تو در تو ترکیبی از **عبارت‌ها در داخل یکدیگر** هستند که به **پردازش‌های پیچیده در یک خط یا بخش مشخص** کمک می‌کنند.  

### **مزایای استفاده از کدهای تو در تو**  
- **افزایش خوانایی و بهینه‌سازی کد**  
- **امکان پردازش چندین مقدار در یک مرحله**  
- **ترکیب عملیات شرطی و محاسبات پیچیده**  

## مقداردهی متغیر (Variable Assignment) در پایتون  

### **مقداردهی متغیر چیست؟**  
مقداردهی متغیر (Variable Assignment) فرآیندی است که طی آن **یک مقدار مشخص** را به **یک نام متغیر** اختصاص می‌دهیم. در پایتون، مقداردهی با استفاده از علامت `=` انجام می‌شود.  

### **ویژگی‌های مقداردهی متغیر در پایتون**  
- **بدون نیاز به تعریف نوع متغیر:** پایتون به‌طور **پویا** نوع متغیر را تشخیص می‌دهد.  
- **امکان مقداردهی چندگانه:** می‌توان چند متغیر را به **یک مقدار مشترک** یا **مقدارهای مختلف** اختصاص داد.  
- **امکان استفاده از مقداردهی موازی:** مقداردهی هم‌زمان چندین متغیر در یک خط انجام می‌شود.  

### **نکات مهم در مقداردهی متغیر**  
✅ مقداردهی **پویا** انجام می‌شود، نیازی به تعیین نوع داده نیست.  
✅ مقداردهی **باید دقیق و واضح** باشد تا از خطاهای منطقی جلوگیری شود.  
✅ نام متغیر **باید معنی‌دار و مرتبط** با کاربرد آن باشد.

# دستور `print()` در پایتون  

## **تعریف `print()`**  
دستور `print()` یکی از پرکاربردترین دستورات در **پایتون** است که برای نمایش خروجی در **کنسول** استفاده می‌شود. این دستور به ما امکان می‌دهد **متغیرها، متن‌ها، نتایج محاسبات و داده‌های دیگر** را روی صفحه چاپ کنیم.  

## **ویژگی‌های دستور `print()`**  
- **نمایش داده‌ها روی خروجی استاندارد (کنسول)**  
- **امکان چاپ چندین مقدار به‌صورت هم‌زمان**  
- **پشتیبانی از قالب‌بندی و تنظیمات دلخواه برای نمایش خروجی**  
- **امکان اضافه کردن جداکننده‌های سفارشی میان داده‌ها**  

## **مثال‌های کاربردی**  
- نمایش متن در کنسول  
- چاپ متغیرها و مقادیر عددی  
- استفاده از جداکننده‌ها و قالب‌بندی پیشرفته  

</div>

```python
original_price = 100
discount_rate = 0.1
tax_rate = 0.08

final_price = (original_price - (original_price * discount_rate)) * (1 + tax_rate)

print("قیمت نهایی محصول:", final_price)
```

# تعریف کامنت (Comment) در پایتون  

## **کامنت چیست؟**  
کامنت (Comment) در پایتون متنی است که داخل کد قرار می‌گیرد اما **توسط مفسر اجرا نمی‌شود**. کامنت‌ها برای **توضیح کد، افزایش خوانایی، و مستندسازی برنامه** استفاده می‌شوند.  

## **مزایای استفاده از کامنت‌ها**  
- توضیح عملکرد بخش‌های مختلف کد  
- کمک به **دیگر برنامه‌نویسان** برای درک بهتر منطق کد  
- افزایش خوانایی و نگهداری بهتر کد  
- مستندسازی پروژه‌های بزرگ  

## **انواع کامنت در پایتون**  
1. **کامنت تک‌خطی** – با `#` شروع می‌شود و فقط در همان خط نوشته می‌شود.  
2. **کامنت چند‌خطی** – با `"""` یا `'''` نوشته می‌شود و چندین خط را شامل می‌شود.  

استفاده صحیح از کامنت‌ها باعث می‌شود کد شما **بهینه‌تر، خواناتر، و قابل‌درک‌تر** باشد! 🚀  


</div>

```python
"""
این بخش از کد محاسبه قیمت نهایی محصول را انجام می‌دهد:
1. ابتدا قیمت تخفیف‌یافته محاسبه می‌شود.
2. سپس مالیات روی قیمت تخفیف‌یافته اضافه می‌شود.
3. در نهایت خروجی چاپ می‌شود.
"""

original_price = 100  # قیمت اولیه محصول
discount_rate = 0.1  # درصد تخفیف
tax_rate = 0.08  # نرخ مالیات

# محاسبه قیمت نهایی پس از اعمال تخفیف و مالیات
final_price = (original_price - (original_price * discount_rate)) * (1 + tax_rate)

print("قیمت نهایی محصول:", final_price)  # نمایش خروجی

```
