# **عبارات شرطی در پایتون (Conditional Statements)**  

## **۱. تعریف عبارات شرطی**  
عبارات شرطی در پایتون برای **کنترل جریان برنامه** استفاده می‌شوند. این دستورات باعث می‌شوند که **برنامه بر اساس شرایط مشخص اجرا شود** و مسیرهای مختلفی را دنبال کند.  

✅ **تصمیم‌گیری بر اساس شرایط** → اجرا یا نادیده‌گیری بخش‌های مختلف کد  
✅ **افزایش کارایی برنامه** → اجرای هوشمندانه‌ی دستورات  
✅ **استفاده از `if`, `elif`, `else`** → تعریف قوانین شرطی  



## **۲. نحوه‌ی نوشتن `if`, `elif`, `else` در پایتون**  

### **🔹 شرط `if`**  
🔹 اگر شرط برقرار باشد، کد اجرا می‌شود.  
🔹 ساختار:  

<div dir='ltr'>


```python
if condition:
    دستورات اجرا شوند
```

</div>


🔹 شرط elif
🔹 بررسی شرط‌های اضافی بعد از if. 
🔹 ساختار:

<div dir='ltr'>

```python

elif condition2:
    دستورات
```

</div>


🔹 شرط else
🔹 اگر هیچ شرطی برقرار نبود، else اجرا می‌شود. 
🔹 ساختار:

<div dir='ltr'>

```python

    دستورات

    دستورات
```

# **White Space (فضای خالی) در پایتون**  

## **۱. تعریف White Space**  
در پایتون، **White Space** به **فاصله‌ها، تورفتگی‌ها (Indentation) و خطوط خالی** اشاره دارد که در ساختار کد تأثیرگذار هستند. برخلاف برخی زبان‌های دیگر، پایتون **از تورفتگی (Indentation) برای مشخص کردن بلوک‌های کد استفاده می‌کند**.  

✅ **تورفتگی جایگزین `{}` در زبان‌هایی مانند C و Java است.**  
✅ **عدم رعایت تورفتگی صحیح باعث خطای `IndentationError` می‌شود.**  
✅ **پایتون معمولاً از ۴ فاصله (Spaces) برای تورفتگی استاندارد استفاده می‌کند.**  
✅ **می‌توان به جای Space از Tab استفاده کرد، اما ترکیب این دو باعث خطا می‌شود.**  



## **۲. کاربردهای White Space در پایتون**  
🔹 **تورفتگی (`Indentation`)** → مشخص کردن بلوک‌های کد مانند حلقه‌ها و شرط‌ها  
🔹 **خطوط خالی (`Blank Lines`)** → خوانایی بهتر کد و جدا کردن بخش‌های مختلف  
🔹 **فاصله بین عملگرها (`Spaces Around Operators`)** → بهبود خوانایی، مانند `x = a + b` به جای `x=a+b`  



## **۳. نمونه‌ی استفاده از تورفتگی صحیح**  
✅ **کد صحیح با تورفتگی استاندارد:**  

<div dir='ltr'>

```python
if True:
    print("این یک کد صحیح است.")  # تورفتگی ۴ فاصله دارد
```

</div>

</div>


````python
# بررسی شرط برای تعیین وضعیت دما
temperature = 25  # مقدار دما

if temperature > 30:
    print("هوا گرم است!")
elif temperature >= 20:
    print("هوا معتدل است.")
else:
    print("هوا سرد است!")

# خروجی برای مقدار 25: "هوا معتدل است."
````

# **حلقه `while` در پایتون**  

## **۱. تعریف حلقه `while`**  
حلقه‌ی `while` در پایتون زمانی استفاده می‌شود که **یک شرط مشخص تا زمانی که برقرار است، اجرا شود**. این حلقه تا زمانی که شرط **`True`** باشد، تکرار خواهد شد.  

✅ **تکرار بر اساس شرط** → تا زمانی که شرط برقرار باشد، کد اجرا می‌شود.  
✅ **عدم نیاز به مقدار مشخصی از تکرار** → مناسب برای شرایط نامشخص.  
✅ **امکان کنترل با دستورات `break` و `continue`** → تنظیم نحوه‌ی اجرای حلقه.  


## **۲. ساختار حلقه `while`**  
🔹 ابتدا **شرط بررسی می‌شود**، اگر `True` باشد، دستورات داخل حلقه اجرا می‌شوند.  
🔹 پس از اجرای دستورات، **شرط دوباره بررسی می‌شود** و در صورت برقرار بودن، حلقه ادامه می‌یابد.  
🔹 در صورتی که شرط `False` شود، اجرای حلقه متوقف خواهد شد.  


<div dir='ltr'>

```python
# حلقه `while` برای شمارش از 1 تا 5
count = 1

while count <= 5:  # تا زمانی که count کوچکتر یا مساوی 5 باشد
    print("مقدار شمارش:", count)
    count += 1  # افزایش مقدار شمارش در هر دور
```

</div>

## ۳. کنترل حلقه با break و continue
🔹 break → خروج فوری از حلقه، حتی اگر شرط برقرار باشد. 
🔹 continue → رد کردن ادامه‌ی اجرای حلقه و رفتن به بررسی شرط مجدد.

<div dir='ltr'>

```python
num = 0

while num < 10:
    num += 1
    if num == 5:
        break  # حلقه هنگام رسیدن به 5 متوقف می‌شود
    print("عدد:", num)
```

```python
num = 0

while num < 5:
    num += 1
    if num == 3:
        continue  # مقدار 3 رد می‌شود و به دور بعدی می‌رود
    print("عدد:", num)
```

</div>


✅ break حلقه را متوقف کرده و از آن خارج می‌شود. 

✅ continue اجرای ادامه‌ی حلقه را رد می‌کند و به بررسی شرط بازمی‌گردد.

حلقه while یکی از مهم‌ترین ساختارهای کنترل جریان در برنامه‌های پایتون است! 🚀



</div>

````python
# برنامه‌ای که از کاربر رمز عبور را دریافت کرده و بررسی می‌کند که آیا درست است یا نه

# تعیین رمز عبور صحیح
correct_password = "Python123"

# دریافت ورودی از کاربر
user_password = input("لطفاً رمز عبور خود را وارد کنید: ")

# بررسی رمز عبور با حلقه `while`
while user_password != correct_password:
    print("رمز عبور اشتباه است! لطفاً دوباره تلاش کنید.")
    user_password = input("لطفاً رمز عبور را وارد کنید: ")

print("ورود موفقیت‌آمیز! خوش آمدید.")
````

# **حلقه `for` در پایتون**  

## **۱. تعریف حلقه `for`**  
حلقه‌ی `for` در پایتون برای **پیمایش مجموعه‌های قابل تکرار** مانند **لیست‌ها، تاپل‌ها، دیکشنری‌ها، رشته‌ها و محدوده‌های عددی (`range`)** استفاده می‌شود. این حلقه به‌طور خودکار **روی عناصر حلقه می‌زند** و نیازی به بررسی شرط ندارد.  

✅ **پیمایش عناصر لیست، تاپل، دیکشنری و رشته‌ها**  
✅ **اجرا بدون نیاز به مقدار مشخصی از تکرار**  
✅ **کنترل اجرای حلقه با `break` و `continue`**  


## **۲. ساختار حلقه `for`**  
🔹 حلقه روی هر عنصر **یکی یکی اجرا می‌شود**.  
🔹 مقدار متغیر درون `for` در هر مرحله **به عنصر بعدی تغییر می‌کند**.  

<div dir='ltr'>

```python
# حلقه `for` برای پیمایش لیستی از اعداد
numbers = [1, 2, 3, 4, 5]

for num in numbers:
    print("عدد:", num)
```

</div>

✅ این حلقه هر مقدار از لیست numbers را چاپ می‌کند و مقدار num در هر مرحله تغییر می‌کند.

۳. استفاده از range() در حلقه for
🔹 تابع range() یک محدوده‌ی عددی ایجاد می‌کند. 
🔹 می‌توان شروع، پایان و گام مشخص کرد.

<div dir='ltr'>

```python
for i in range(1, 6):  # از ۱ تا ۵
    print("عدد:", i)

```

</div>

۴. پیمایش رشته‌ها با حلقه for
🔹 می‌توان هر کاراکتر رشته را جداگانه پردازش کرد.

<div dir='ltr'>

```python
text = "Python"

for char in text:
    print("کاراکتر:", char)

```

</div>

۵. کنترل حلقه با break و continue
🔹 break → خروج از حلقه، حتی اگر هنوز عناصر باقی باشند. 
🔹 continue → رد کردن مرحله‌ی فعلی و رفتن به عنصر بعدی.

<div dir='ltr'>

```python
for i in range(1, 6):
    if i == 3:
        break  # حلقه هنگام رسیدن به مقدار ۳ متوقف می‌شود
    print("عدد:", i)


for i in range(1, 6):
    if i == 3:
        continue  # مقدار ۳ رد می‌شود و اجرا نمی‌شود
    print("عدد:", i)


```

</div>

✅ break حلقه را فوراً متوقف می‌کند، در حالی که continue فقط مقدار جاری را نادیده می‌گیرد.

</div>

````python
# محاسبه مجموع اعداد زوج در یک لیست با استفاده از حلقه `for`

# تعریف لیست اعداد
numbers = [12, 7, 9, 14, 6, 21, 30, 17]

# متغیر برای ذخیره مجموع اعداد زوج
sum_even = 0

# حلقه `for` برای بررسی هر عدد در لیست
for num in numbers:
    if num % 2 == 0:  # بررسی عدد زوج
        sum_even += num  # اضافه کردن عدد زوج به مجموع

# چاپ مجموع اعداد زوج
print("مجموع اعداد زوج:", sum_even)
````
