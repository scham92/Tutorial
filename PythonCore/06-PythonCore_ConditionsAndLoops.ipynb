# **عبارات شرطی در پایتون (Conditional Statements)**  

## **۱. تعریف عبارات شرطی**  
عبارات شرطی در پایتون برای **کنترل جریان برنامه** استفاده می‌شوند. این دستورات باعث می‌شوند که **برنامه بر اساس شرایط مشخص اجرا شود** و مسیرهای مختلفی را دنبال کند.  

✅ **تصمیم‌گیری بر اساس شرایط** → اجرا یا نادیده‌گیری بخش‌های مختلف کد  
✅ **افزایش کارایی برنامه** → اجرای هوشمندانه‌ی دستورات  
✅ **استفاده از `if`, `elif`, `else`** → تعریف قوانین شرطی  



## **۲. نحوه‌ی نوشتن `if`, `elif`, `else` در پایتون**  

### **🔹 شرط `if`**  
🔹 اگر شرط برقرار باشد، کد اجرا می‌شود.  
🔹 ساختار:  

<div dir='ltr'>


```python
if condition:
    دستورات اجرا شوند
```

</div>


🔹 شرط elif
🔹 بررسی شرط‌های اضافی بعد از if. 
🔹 ساختار:

<div dir='ltr'>

```python
if condition1:
    دستورات
elif condition2:
    دستورات
```

</div>


🔹 شرط else
🔹 اگر هیچ شرطی برقرار نبود، else اجرا می‌شود. 
🔹 ساختار:

<div dir='ltr'>

```python
if condition:
    دستورات
else:
    دستورات
```

</div>

# **White Space (فضای خالی) در پایتون**  

## **۱. تعریف White Space**  
در پایتون، **White Space** به **فاصله‌ها، تورفتگی‌ها (Indentation) و خطوط خالی** اشاره دارد که در ساختار کد تأثیرگذار هستند. برخلاف برخی زبان‌های دیگر، پایتون **از تورفتگی (Indentation) برای مشخص کردن بلوک‌های کد استفاده می‌کند**.  

✅ **تورفتگی جایگزین `{}` در زبان‌هایی مانند C و Java است.**  
✅ **عدم رعایت تورفتگی صحیح باعث خطای `IndentationError` می‌شود.**  
✅ **پایتون معمولاً از ۴ فاصله (Spaces) برای تورفتگی استاندارد استفاده می‌کند.**  
✅ **می‌توان به جای Space از Tab استفاده کرد، اما ترکیب این دو باعث خطا می‌شود.**  



## **۲. کاربردهای White Space در پایتون**  
🔹 **تورفتگی (`Indentation`)** → مشخص کردن بلوک‌های کد مانند حلقه‌ها و شرط‌ها  
🔹 **خطوط خالی (`Blank Lines`)** → خوانایی بهتر کد و جدا کردن بخش‌های مختلف  
🔹 **فاصله بین عملگرها (`Spaces Around Operators`)** → بهبود خوانایی، مانند `x = a + b` به جای `x=a+b`  



## **۳. نمونه‌ی استفاده از تورفتگی صحیح**  
✅ **کد صحیح با تورفتگی استاندارد:**  

<div dir='ltr'>

```python
if True:
    print("این یک کد صحیح است.")  # تورفتگی ۴ فاصله دارد
```

</div>

</div>


````python
# بررسی شرط برای تعیین وضعیت دما
temperature = 25  # مقدار دما

if temperature > 30:
    print("هوا گرم است!")
elif temperature >= 20:
    print("هوا معتدل است.")
else:
    print("هوا سرد است!")

# خروجی برای مقدار 25: "هوا معتدل است."
````

# **حلقه `while` در پایتون**  

## **۱. تعریف حلقه `while`**  
حلقه‌ی `while` در پایتون زمانی استفاده می‌شود که **یک شرط مشخص تا زمانی که برقرار است، اجرا شود**. این حلقه تا زمانی که شرط **`True`** باشد، تکرار خواهد شد.  

✅ **تکرار بر اساس شرط** → تا زمانی که شرط برقرار باشد، کد اجرا می‌شود.  
✅ **عدم نیاز به مقدار مشخصی از تکرار** → مناسب برای شرایط نامشخص.  
✅ **امکان کنترل با دستورات `break` و `continue`** → تنظیم نحوه‌ی اجرای حلقه.  

---

## **۲. ساختار حلقه `while`**  
🔹 ابتدا **شرط بررسی می‌شود**، اگر `True` باشد، دستورات داخل حلقه اجرا می‌شوند.  
🔹 پس از اجرای دستورات، **شرط دوباره بررسی می‌شود** و در صورت برقرار بودن، حلقه ادامه می‌یابد.  
🔹 در صورتی که شرط `False` شود، اجرای حلقه متوقف خواهد شد.  

<div dir='rtl'>

```python
# حلقه `while` برای شمارش از 1 تا 5
count = 1

while count <= 5:  # تا زمانی که count کوچکتر یا مساوی 5 باشد
    print("مقدار شمارش:", count)
    count += 1  # افزایش مقدار شمارش در هر دور
```

</div>

## ۳. کنترل حلقه با break و continue
🔹 break → خروج فوری از حلقه، حتی اگر شرط برقرار باشد. 
🔹 continue → رد کردن ادامه‌ی اجرای حلقه و رفتن به بررسی شرط مجدد.

<div dir='ltr'>

```python
num = 0

while num < 10:
    num += 1
    if num == 5:
        break  # حلقه هنگام رسیدن به 5 متوقف می‌شود
    print("عدد:", num)
```

```python
num = 0

while num < 5:
    num += 1
    if num == 3:
        continue  # مقدار 3 رد می‌شود و به دور بعدی می‌رود
    print("عدد:", num)
```

</div>


✅ break حلقه را متوقف کرده و از آن خارج می‌شود. 
✅ continue اجرای ادامه‌ی حلقه را رد می‌کند و به بررسی شرط بازمی‌گردد.

حلقه while یکی از مهم‌ترین ساختارهای کنترل جریان در برنامه‌های پایتون است! 🚀
