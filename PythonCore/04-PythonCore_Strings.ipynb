# رشته‌ها (Strings) در پایتون  

## **تعریف رشته (String) در پایتون**  
رشته (String) یک نوع داده **متنی** در پایتون است که شامل **کاراکترها، کلمات، یا جملات** می‌شود. رشته‌ها داخل **علامت نقل‌قول (`"..."` یا `'...'`)** قرار می‌گیرند و برای **ذخیره و پردازش داده‌های متنی** استفاده می‌شوند.  

## **ویژگی‌های رشته در پایتون**  
✅ **قابل تغییر نیستند** (Immutable)؛ یعنی بعد از تعریف نمی‌توان مقدار آن را تغییر داد.  
✅ می‌توان **از نقل‌قول تکی (`'...'`) یا دوتایی (`"..."`)** استفاده کرد.  
✅ امکان **استفاده از رشته‌های چندخطی** با `"""` یا `'''` وجود دارد.  

## **روش‌های تعریف رشته‌ها**  

</div>

```python
text1 = "Hello, Python!"  # رشته با نقل‌قول دوتایی
text2 = 'Hello, AI!'      # رشته با نقل‌قول تکی
text3 = """این یک رشته
 چندخطی است."""  # رشته چندخطی

print(text1)
print(text2)
print(text3)
```

# عملیات جمع (`+`) و ضرب (`*`) روی رشته‌ها در پایتون  

## **۱. عملیات جمع (`+`) روی رشته‌ها**  
عملگر `+` برای **اتصال دو رشته** به یکدیگر استفاده می‌شود و یک رشته‌ی جدید تولید می‌کند.  

**مثال:**  

</div>

```python
first_name = "Sajad"
last_name = "Chamanara"
full_name = first_name + " " + last_name
print(full_name)
```

<div dir='rtl'>


## **۲. عملیات ضرب (`*`) روی رشته‌ها**  
عملگر `*` باعث تکرار یک رشته به تعداد مشخص می‌شود.  

**نکته:** عدد ضرب شده باید مثبت و صحیح باشد.

**مثال:**  


</div>


```python
text = "Python "
repeated_text = text * 3
print(repeated_text)
```

# **ایندکس در رشته‌ها و متغیرهای آرایه‌ای در پایتون**  

## **۱. ایندکس در رشته‌ها (String Indexing)**  
رشته‌ها در پایتون مانند **یک آرایه از کاراکترها** عمل می‌کنند، و هر کاراکتر دارای **ایندکس** مخصوص به خود است.  
- ایندکس از **0** شروع می‌شود و هر کاراکتر یک شماره دارد.  
- امکان **استفاده از ایندکس منفی** وجود دارد که از انتهای رشته شمارش می‌شود (`-1` برای کاراکتر آخر).  
- قابلیت **برش (Slicing)** برای دریافت بخشی از رشته بر اساس ایندکس‌ها.  

## **۲. ایندکس در لیست‌ها و تاپل‌ها (Lists & Tuples Indexing)**  
لیست‌ها و تاپل‌ها نیز **دارای ایندکس** هستند که امکان دسترسی به **عناصر مشخص** را فراهم می‌کند.  
- مقدار هر عنصر با **ایندکس عددی** قابل دریافت است.  
- **ایندکس منفی** امکان دریافت مقدار از انتهای لیست یا تاپل را دارد (`-1` برای آخرین مقدار، `-2` برای مقدار قبل از آن).  
- قابلیت **برش (Slicing)** برای گرفتن چندین مقدار به‌صورت هم‌زمان.  

### **ویژگی‌های ایندکس و برش در پایتون:**  
✅ **دسترسی سریع** به داده‌ها در رشته‌ها و لیست‌ها.  
✅ **مدیریت و پردازش داده‌ها** با انعطاف‌پذیری بالا.  
✅ **امکان شمارش از ابتدا یا انتها** با استفاده از ایندکس‌های مثبت و منفی.  

ایندکس و برش یکی از ابزارهای قدرتمند برای **مدیریت داده‌های متنی و عددی** در پایتون هستند! 🚀  


</div>

```python
# تعریف یک رشته نمونه
text = "Python Programming"

# دسترسی به کاراکتر خاص با ایندکس
print("اولین کاراکتر:", text[0])  
print("آخرین کاراکتر:", text[-1])  

# برش (Slicing) بخش‌های مختلف رشته
print("از ایندکس 0 تا 5:", text[0:6])  
print("از ایندکس 7 تا آخر:", text[7:])  
print("از ابتدا تا ایندکس 5:", text[:6])  

# استفاده از ایندکس منفی در برش رشته
print("از ایندکس منفی 11 تا منفی 2:", text[-11:-1])  

# برش با گام (Step) مشخص
print("گرفتن هر کاراکتر دوم:", text[::2])  
print("معکوس کردن رشته:", text[::-1])  
```

# **متدها در رشته‌های پایتون**  

## **۱. تعریف متد در پایتون**  
متدها در پایتون **توابع داخلی** هستند که برای **پردازش و تغییر رشته‌ها** استفاده می‌شوند. این متدها می‌توانند **مقدار جدیدی ایجاد کنند، تغییرات اعمال کنند، یا اطلاعات خاصی از رشته استخراج کنند**.  

## **۲. متدهای پرکاربرد در رشته‌ها**  

### **الف) متدهای تغییر و اصلاح متن**  
- **`upper()`** → تبدیل تمام کاراکترهای رشته به **حروف بزرگ**  
- **`lower()`** → تبدیل تمام کاراکترهای رشته به **حروف کوچک**  
- **`strip()`** → حذف فضاهای خالی از **ابتدا و انتهای** رشته  
- **`replace(old, new)`** → جایگزینی یک زیررشته با مقدار جدید  
- **`capitalize()`** → بزرگ کردن اولین حرف رشته  

### **ب) متدهای بررسی و جستجو در رشته**  
- **`startswith(substring)`** → بررسی **آغاز رشته** با یک زیررشته مشخص  
- **`endswith(substring)`** → بررسی **پایان رشته** با یک زیررشته مشخص  
- **`find(substring)`** → یافتن **موقعیت اولین وقوع** یک زیررشته در متن  
- **`count(substring)`** → شمارش تعداد وقوع یک زیررشته در رشته  

### **ج) متدهای تقسیم و ترکیب رشته‌ها**  
- **`split(separator)`** → تقسیم رشته به **لیستی از بخش‌ها** با استفاده از یک جداکننده  
- **`join(iterable)`** → ترکیب مجموعه‌ای از رشته‌ها با یک جداکننده مشخص  

### **د) متدهای بررسی محتوا**  
- **`isnumeric()`** → بررسی اینکه **رشته فقط شامل اعداد است**  
- **`isalpha()`** → بررسی اینکه **رشته فقط شامل حروف الفبا است**  
- **`isalnum()`** → بررسی اینکه **رشته فقط شامل حروف و اعداد است**  

متدهای رشته‌ای در پایتون ابزارهای قدرتمندی برای **مدیریت، پردازش و تحلیل داده‌های متنی** هستند! 🚀  

</div>


```python
# مثال‌های متدهای رشته‌ای در پایتون

# ۱. تبدیل متن به حروف بزرگ و کوچک
text = "Python Programming"
print("متن اصلی:", text)
print("حروف بزرگ:", text.upper())  # تبدیل به حروف بزرگ
print("حروف کوچک:", text.lower())  # تبدیل به حروف کوچک

# ۲. حذف فضاهای خالی از ابتدا و انتهای رشته
text_with_spaces = "   Hello, World!   "
print("متن اصلی با فضای اضافی:", text_with_spaces)
print("بعد از حذف فضاها:", text_with_spaces.strip())  # حذف فضاها

# ۳. جایگزینی بخشی از متن
message = "I love Java"
print("متن اصلی:", message)
print("بعد از جایگزینی:", message.replace("Java", "Python"))  # تغییر Java به Python

# ۴. بررسی شروع و پایان رشته
filename = "report.pdf"
print("آیا فایل با 'report' شروع می‌شود؟", filename.startswith("report"))  # بررسی شروع
print("آیا فایل با '.pdf' پایان می‌یابد؟", filename.endswith(".pdf"))  # بررسی پایان

# ۵. جستجو و شمارش زیررشته‌ها
text_sample = "Python is powerful and Python is popular"
print("موقعیت اولین 'Python' در متن:", text_sample.find("Python"))  # موقعیت اولین Python
print("تعداد دفعات 'Python' در متن:", text_sample.count("Python"))  # شمارش تعداد Python

# ۶. تقسیم رشته و ترکیب مجدد
csv_data = "apple,banana,grape"
split_data = csv_data.split(",")  # تقسیم رشته بر اساس ","
print("تقسیم‌شده:", split_data)
print("ترکیب مجدد:", "-".join(split_data))  # اتصال لیست با "-"

# ۷. بررسی نوع داده‌های داخل رشته
numeric_text = "12345"
alpha_text = "Python"
mixed_text = "Python123"

print("آیا فقط عدد است؟", numeric_text.isnumeric())  # بررسی اعداد
print("آیا فقط حروف است؟", alpha_text.isalpha())  # بررسی حروف
print("آیا فقط شامل حروف و عدد است؟", mixed_text.isalnum())  # بررسی ترکیب حروف و عدد
```

# **جدول متدهای رشته‌ای در پایتون**  

متدهای رشته‌ای در پایتون برای **ویرایش، پردازش و جستجوی** داده‌های متنی استفاده می‌شوند. جدول زیر **نام متد، ساختار کد، نتیجه و توضیح کاربردی** هر متد را نمایش می‌دهد:  

<div dir='ltr'; align='center'>

| **نام متد**    | **ساختار کد**               | **نتیجه**                  | **توضیح** |
|--------------|-------------------------|--------------------------|------------|
| **`upper()`** | `"hello".upper()`        | `"HELLO"`                | تبدیل تمام کاراکترهای رشته به **حروف بزرگ** |
| **`lower()`** | `"HELLO".lower()`        | `"hello"`                | تبدیل تمام کاراکترهای رشته به **حروف کوچک** |
| **`capitalize()`** | `"python".capitalize()` | `"Python"`               | بزرگ کردن **اولین حرف رشته** |
| **`title()`** | `"hello world".title()`  | `"Hello World"`          | تبدیل اولین حرف **هر کلمه** به **حروف بزرگ** |
| **`strip()`** | `"  hello  ".strip()`    | `"hello"`                | حذف **فضاهای خالی** از ابتدا و انتهای رشته |
| **`lstrip()`** | `"  hello".lstrip()`    | `"hello"`                | حذف **فضاهای خالی** از ابتدای رشته |
| **`rstrip()`** | `"hello  ".rstrip()`    | `"hello"`                | حذف **فضاهای خالی** از انتهای رشته |
| **`replace()`** | `"hello".replace("h", "H")` | `"Hello"`                | جایگزینی یک **زیررشته** با مقدار جدید |
| **`startswith()`** | `"hello".startswith("h")` | `True`                  | بررسی **آغاز رشته** با یک مقدار مشخص |
| **`endswith()`** | `"hello".endswith("o")` | `True`                  | بررسی **پایان رشته** با یک مقدار مشخص |
| **`find()`** | `"hello".find("l")`      | `2`                      | یافتن **موقعیت اولین وقوع** یک زیررشته |
| **`count()`** | `"hello".count("l")`    | `2`                      | شمارش **تعداد وقوع** یک زیررشته |
| **`split()`** | `"apple,banana".split(",")` | `['apple', 'banana']`   | **تقسیم رشته** به لیست با استفاده از جداکننده |
| **`join()`** | `" - ".join(["a", "b"])` | `"a - b"`               | **ترکیب رشته‌ها** با استفاده از جداکننده |
| **`isnumeric()`** | `"123".isnumeric()`   | `True`                   | بررسی اینکه **رشته فقط شامل عدد است** |
| **`isalpha()`** | `"hello".isalpha()`    | `True`                   | بررسی اینکه **رشته فقط شامل حروف است** |
| **`isalnum()`** | `"hello123".isalnum()` | `True`                   | بررسی اینکه **رشته شامل حروف و عدد است** |

</div>

این جدول شامل **متدهای پرکاربرد رشته‌ای** در پایتون است که به **مدیریت و پردازش داده‌های متنی** کمک می‌کند. 🚀
