# انواع داده‌ها (Data Types) در پایتون  

پایتون دارای **انواع داده‌های متنوع** برای ذخیره و پردازش اطلاعات است. این داده‌ها شامل **اعداد، متن‌ها، لیست‌ها، تاپل‌ها، دیکشنری‌ها، مجموعه‌ها و داده‌های منطقی** هستند. در جدول زیر انواع داده‌های اصلی در پایتون به همراه توضیح ارائه شده است:  

<div align='center'>

| **نوع داده**   | **نمونه**          | **توضیح** |
|---------------|------------------|-----------|
| **عدد صحیح (int)** | `x = 10` | شامل اعداد بدون اعشار |
| **عدد اعشاری (float)** | `x = 3.14` | شامل اعداد دارای اعشار |

# دستور `type()` در پایتون  

## **تعریف `type()`**  
دستور `type()` در پایتون برای **تعیین نوع داده** یک متغیر یا مقدار استفاده می‌شود. این دستور **نوع داده‌ای متغیر را برمی‌گرداند** و به برنامه‌نویسان کمک می‌کند تا نوع داده‌های ذخیره‌شده را بررسی کنند.  

## **ویژگی‌های `type()`**  
✅ تعیین **نوع داده‌ی عددی، رشته‌ای، لیست، دیکشنری و غیره**  
✅ کمک به **درک ساختار داده‌ها** در برنامه‌های پیچیده  
✅ امکان **استفاده در تبدیل و اعتبارسنجی داده‌ها**  

## **کاربردهای `type()`**  
- بررسی **نوع متغیرها** قبل از انجام عملیات روی آن‌ها  
- **اشکال‌زدایی** و رفع مشکلات مربوط به نوع داده‌ها  
- استفاده در **تبدیل انواع داده‌ها** برای سازگاری بهتر  


</div>

```python
x = 10
print(type(x))  
```

```python
y = 3.14
print(type(y))  
```

```python
dict_data = {"name": "Sajad", "age": 25}
print(type(dict_data))  
```

# انواع داده‌های عددی و تبدیل آن‌ها در پایتون  

## **۱. تعریف اعداد صحیح (Integer - `int`)**  
اعداد صحیح شامل **تمام اعداد بدون اعشار** هستند، مانند `5`، `-10`، `200`. این داده‌ها برای **محاسبات عددی، شمارش و پردازش بدون نیاز به دقت اعشاری** استفاده می‌شوند.  

## **۲. تعریف اعداد اعشاری (Floating Point - `float`)**  
اعداد اعشاری شامل **اعداد دارای بخش اعشاری** هستند، مانند `3.14`، `-0.5`، `100.75`. این داده‌ها برای **محاسبات دقیق‌تر** و **محاسبات علمی و مالی** مناسب هستند.  

## **۳. تبدیل اعداد صحیح به اعشاری**  
برای تبدیل یک **عدد صحیح (`int`)** به **عدد اعشاری (`float`)** از تابع `float()` استفاده می‌شود.  

**مثال:**  

</div>

```python
x = 10
y = float(x)
print(y)
print('type x: ',type(x))
print('type y: ',type(y))
```

# عملیات (Operations) در پایتون  

عملیات در پایتون شامل **عملگرهای ریاضی، منطقی، مقایسه‌ای و انتسابی** است که برای پردازش داده‌ها استفاده می‌شوند. در جدول زیر انواع عملیات ریاضی و کاربرد آن‌ها ارائه شده است:  

<div align='center'>

| **نوع عملیات** | **نمونه** | **توضیح** |

| **جمع (+)** | `x + y` | مجموع دو مقدار |

# مثال حل معادله درجه دوم و یافتن ریشه‌ها در پایتون  

معادله درجه دوم به‌صورت کلی به شکل زیر تعریف می‌شود:  

$$ ax^2 + bx + c = 0 $$  

برای یافتن ریشه‌های این معادله، از **فرمول دلتا** استفاده می‌کنیم:  

$$ x = \frac{-b \pm \sqrt{\Delta}}{2a} $$  

که در آن:  

$$ \Delta = b^2 - 4ac $$  

### **کد برای حل معادله درجه دوم** 

</div>

```python
# ضرایب معادله درجه دوم
a = 1
b = -3
c = 2

# محاسبه دلتا
delta = b**2 - 4*a*c

# بررسی مقدار دلتا برای تعیین تعداد ریشه‌ها
if delta > 0:
    x1 = (-b + delta**2) / (2*a)
    x2 = (-b - delta**2) / (2*a)
    print("ریشه‌ها:", x1, "و", x2)
elif delta == 0:
    x = -b / (2*a)
    print("ریشه‌ی یکتا:", x)
else:
    print("معادله ریشه‌ی حقیقی ندارد.")
```

# داده‌های بولی (Boolean) در پایتون  

**داده‌های بولی** یکی از انواع داده‌های پایه در پایتون هستند که فقط می‌توانند دو مقدار داشته باشند:  
✅ **`True`** (درست)  
❌ **`False`** (نادرست)  

این داده‌ها برای **شرط‌گذاری، بررسی وضعیت‌ها، و تصمیم‌گیری در کد** استفاده می‌شوند. مقدار `True` به معنای **فعال بودن شرط** و مقدار `False` به معنای **غیرفعال بودن شرط** است.  


# عملگرهای مقایسه‌ای (Comparison Operators)  

عملگرهای مقایسه‌ای در پایتون برای **مقایسه مقدارها** استفاده می‌شوند و مقدار `True` یا `False` تولید می‌کنند.  

<div align='center'>

| **عملگر** | **نمونه** | **توضیح** |
|-----------|----------|-----------|
| **`==`** | `x == y` | بررسی **برابری** مقدار دو متغیر |
| **`!=`** | `x != y` | بررسی **نابرابری** مقدار دو متغیر |
| **`>`** | `x > y` | بررسی **بزرگ‌تر بودن** مقدار `x` نسبت به `y` |
| **`<`** | `x < y` | بررسی **کوچک‌تر بودن** مقدار `x` نسبت به `y` |
| **`>=`** | `x >= y` | بررسی **بزرگ‌تر یا مساوی** بودن مقدار `x` نسبت به `y` |
| **`<=`** | `x <= y` | بررسی **کوچک‌تر یا مساوی** بودن مقدار `x` نسبت به `y` |

</div>

✅ این عملگرها معمولاً در **دستورات شرطی (`if`)** و بررسی وضعیت‌ها استفاده می‌شوند.  


# عملگرهای منطقی (Logical Operators)  

عملگرهای منطقی برای **ترکیب چندین شرط** و بررسی وضعیت‌ها استفاده می‌شوند.  

<div align='center'>

| **عملگر** | **نمونه** | **توضیح** |
|-----------|----------|-----------|
| **`and`** | `x > 5 and y < 10` | اگر هر دو شرط **True** باشند، خروجی **True** است. |
| **`or`** | `x > 5 or y < 10` | اگر حداقل یکی از شرط‌ها **True** باشد، خروجی **True** است. |
| **`not`** | `not x > 5` | نتیجه‌ی عبارت را معکوس می‌کند (`True` → `False` و برعکس). |

</div>

✅ این عملگرها برای **تصمیم‌گیری پیچیده و بررسی چندین وضعیت به‌طور هم‌زمان** مفید هستند.  

با ترکیب **عملگرهای مقایسه‌ای و منطقی** می‌توان شرط‌های پیچیده را در برنامه‌نویسی مدیریت کرد! 🚀  

</div>


```python
age = 25

print(age > 18)   # خروجی: True
print(age < 30)   # خروجی: True
print(age == 25)  # خروجی: True
print(age != 20)  # خروجی: True
print(age >= 25)  # خروجی: True
print(age <= 18)  # خروجی: False
```

```python
score = 85
passed = score >= 60
high_score = score > 90

print(passed and high_score)  # خروجی: False
print(passed or high_score)   # خروجی: True
print(not passed)             # خروجی: False
```

```python
temperature = 25
is_warm = temperature > 20 and temperature < 30
is_cold = temperature < 15 or temperature > 35

print(is_warm)  # خروجی: True
print(is_cold)  # خروجی: False
```
