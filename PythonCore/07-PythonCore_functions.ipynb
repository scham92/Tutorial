# **توابع (Functions) در پایتون**  

## **۱. تعریف توابع**  
توابع در پایتون مجموعه‌ای از **دستورات قابل استفاده‌ی مجدد** هستند که برای **کاهش کدهای تکراری، بهبود خوانایی و افزایش کارایی برنامه** استفاده می‌شوند. با تعریف یک تابع، می‌توان آن را **چندین بار** فراخوانی و اجرا کرد.  

✅ **کاهش تکرار کد** → امکان استفاده‌ی مجدد از دستورات.  
✅ **ساختار منظم‌تر** → افزایش خوانایی و مدیریت بهتر کد.  
✅ **امکان دریافت ورودی (`parameters`) و بازگردانی مقدار (`return`).**  



## **۲. نحوه‌ی تعریف یک تابع**  
🔹 توابع با استفاده از **کلمه‌ی کلیدی `def`** تعریف می‌شوند.  
🔹 باید یک **نام مناسب برای تابع** انتخاب شود.  
🔹 می‌توان **ورودی‌ها (`parameters`)** را مشخص کرد (اختیاری).  
🔹 می‌توان مقدار **بازگشتی (`return`)** تعیین کرد (اختیاری).  

<div dir='ltr'>

```python
# تعریف یک تابع برای محاسبه مجموع دو عدد
def add_numbers(a, b):
    return a + b  # بازگردانی مجموع

# فراخوانی تابع
result = add_numbers(5, 3)
print("مجموع:", result)
```

</div>


✅ تابع add_numbers دو عدد را دریافت کرده و مجموع آن‌ها را برمی‌گرداند. 
✅ با return مقدار به متغیر result منتقل شده و نمایش داده می‌شود.

## ۳. انواع توابع در پایتون
🔹 تابع بدون ورودی و خروجی → فقط مجموعه‌ای از دستورات را اجرا می‌کند. 
🔹 تابع با ورودی (parameters) → مقدارهای مختلف را دریافت کرده و پردازش می‌کند. 
🔹 تابع با خروجی (return) → مقدار پردازش شده را بازمی‌گرداند. 
🔹 تابع درونی (lambda) → توابع کوچک و ساده بدون نام.

</div>


````python
# تابعی برای تبدیل واحد دما از سلسیوس به فارنهایت و بالعکس

def convert_temperature(value, unit):
    if unit.lower() == "c":  # تبدیل از سلسیوس به فارنهایت
        return (value * 9/5) + 32
    elif unit.lower() == "f":  # تبدیل از فارنهایت به سلسیوس
        return (value - 32) * 5/9
    else:
        return "واحد نامعتبر! لطفاً 'C' یا 'F' وارد کنید."

# دریافت ورودی از کاربر
temperature = float(input("مقدار دما را وارد کنید: "))
unit = input("واحد دما (C یا F): ")

# فراخوانی تابع و نمایش نتیجه
converted_temp = convert_temperature(temperature, unit)
print("مقدار تبدیل‌شده:", converted_temp)
````

# **توابع Lambda در پایتون**  

## **۱. تعریف تابع Lambda**  
تابع `lambda` یک **تابع ناشناس (بدون نام)** و کوتاه در پایتون است که **در یک خط تعریف می‌شود** و معمولاً برای عملیات‌های ساده و سریع استفاده می‌شود.  

✅ **ساختار ساده و کوتاه** → بدون نیاز به `def` و بلوک‌های چندخطی.  
✅ **عملکرد سریع در موارد ساده** → مناسب برای توابعی که فقط یک بار استفاده می‌شوند.  
✅ **استفاده در توابع `map()`, `filter()`, و `reduce()`** → کاربرد زیاد در پردازش داده.  


## **۲. تفاوت Lambda با تابع `def`**  

<div dir='ltr' align='center'>

| **ویژگی**           | **تابع `lambda`**  | **تابع `def`** |
|------------------|----------------|---------------|
| **نام تابع**    | بدون نام (ناشناس) | دارای نام مشخص |
| **تعداد خطوط**  | در یک خط تعریف می‌شود | می‌تواند شامل چندین خط باشد |
| **انعطاف‌پذیری** | محدود به یک عبارت | امکان استفاده از چندین دستور |
| **خوانایی کد**  | ساده و کوتاه | خواناتر برای توابع پیچیده |
| **محدودیت‌ها**  | نمی‌توان `return`، `loop` یا `try` استفاده کرد | امکان تعریف انواع ساختارها |
| **کاربرد اصلی** | در `map()`, `filter()`, و محاسبات سریع | توابع پیچیده و سازمان‌دهی شده |

</div>



## **۳. نحوه‌ی نوشتن Lambda**  
🔹 ساختار **`lambda ورودی‌ها: خروجی`** است.  
🔹 می‌توان چندین آرگومان داشت، اما فقط **یک عبارت** را اجرا می‌کند.  

<div dir='ltr'>

```python
# تابع Lambda برای جمع دو عدد
sum_lambda = lambda a, b: a + b
print(sum_lambda(5, 3))  # خروجی: 8
```

</div>

✅ تابع lambda یک خطی است و به‌طور مستقیم مقدار را برمی‌گرداند. 
✅ نیازی به return ندارد، زیرا خروجی خودکار تولید می‌شود.

## ۴. چه زمانی از Lambda استفاده کنیم؟
🔹 در توابعی که فقط یک بار استفاده می‌شوند. 
🔹 در مواردی که نیاز به نوشتن کد کوتاه و سریع داریم. 
🔹 برای پردازش داده‌ها با map(), filter(), reduce(). 
🔹 درون لیست‌ها و دیکشنری‌ها برای پردازش سریع اطلاعات.

<div dir='ltr'>

```python
# استفاده از Lambda در filter()
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # خروجی: [2, 4, 6]
```

</div>

✅ تابع lambda به‌طور مستقیم مقدارهای زوج را فیلتر کرده است.

# **۵. مزایا و معایب Lambda**  

<div dir='ltr' align='center'>

| **مزایا**                     | **معایب** |
|----------------------------|---------|
| کد کوتاه و خوانا                | محدود به یک عبارت و عدم امکان استفاده از چندین دستور |
| اجرا سریع در پردازش‌های ساده       | قابلیت دیباگ کردن سخت‌تر است |
| مناسب برای استفاده در `map()`, `filter()`, و `reduce()` | خوانایی کم در توابع پیچیده |
| مناسب برای استفاده موقت         | عدم امکان تعریف متدهای کلاس |

</div>

✅ **توابع `lambda` سریع، کوتاه و مناسب برای پردازش‌های ساده هستند، اما توابع `def` برای کدهای پیچیده مناسب‌ترند!** 🚀  




</div>

````python
# تبدیل لیست رشته‌ها به حروف بزرگ با استفاده از `lambda` و `map()`

# لیست رشته‌ها
words = ["python", "data", "science", "machine learning"]

# استفاده از `lambda` برای تبدیل تمامی کلمات به حروف بزرگ
uppercase_words = list(map(lambda word: word.upper(), words))

# نمایش لیست جدید
print(uppercase_words)
````

# **۱. متغیرهای محلی (`Local Variables`)**  
✅ فقط درون **تابع** قابل دسترسی هستند و پس از خروج از تابع **از بین می‌روند**.  
✅ مناسب برای ذخیره اطلاعات **موقت** و **جلوگیری از تغییر ناخواسته‌ی داده‌های اصلی**.  

<div dir='ltr'>

```python
def my_function():

```

# **۲. متغیرهای سراسری (`Global Variables`)**  
✅ **خارج از تابع** تعریف می‌شوند و در **تمام قسمت‌های برنامه** قابل دسترسی هستند.  
✅ برای ذخیره **تنظیمات کلی** یا داده‌هایی که در چندین قسمت پروژه نیاز است، مفید هستند.  


<div dir='ltr'>

```python
global_var = "Hello"

def my_function():

```

</div>


# **۳. استفاده از `global` درون تابع**  
✅ اگر داخل تابع بخواهیم **مقدار یک متغیر سراسری را تغییر دهیم**، باید از `global` استفاده کنیم.  


<div dir='ltr'>

```python
counter = 0

def increase():
    global counter  # اعلام می‌کنیم که این متغیر در سطح جهانی تغییر می‌کند
    counter += 1

increase()
print(counter)  # خروجی: 1
```

</div>


# **۴. متغیرهای غیرمحلی (`nonlocal`)**  
✅ داخل **توابع داخلی (Nested Functions)** استفاده می‌شوند.  
✅ مقدارشان فقط در سطح تابع بالادستی تغییر می‌کند، نه در سطح سراسری.  


<div dir='ltr'>

```python
def outer_function():
    x = 5
    def inner_function():
        nonlocal x  # مقدار متغیر x درون تابع بالایی تغییر می‌کند
        x = 10
    inner_function()
    print(x)  # خروجی: 10
```

</div>

</div>

````python
# مدیریت مقدار فروش با استفاده از متغیرهای محلی، گلوبال و غیرمحلی

# متغیر سراسری برای نگه‌داری کل فروش
total_sales = 0

def update_sales():
    global total_sales  # تغییر مقدار متغیر سراسری
    total_sales += 100  # افزایش فروش
    total_salesM = total_sales

    def daily_sales():
        nonlocal total_salesM  # استفاده از مقدار متغیر تابع خارجی
        local_sales = 50  # متغیر محلی فقط داخل تابع داخلی قابل دسترسی است
        total_salesM += local_sales  # اضافه کردن فروش روزانه به کل فروش
        print("فروش روزانه:", local_sales)

    daily_sales()
    print("مجموع فروش تا کنون:", total_sales)

# اجرای تابع
update_sales()
print("مجموع فروش نهایی:", total_sales)
````

# **۱. تعریف `try` و `except` در پایتون**  
عبارت **`try-except`** در پایتون برای **مدیریت خطاها و جلوگیری از توقف ناگهانی برنامه** استفاده می‌شود. این ساختار **اجرای ایمن کد** را تضمین می‌کند و به جای نمایش خطای برنامه، امکان **واکنش مناسب** به خطا را فراهم می‌آورد.  

✅ **جلوگیری از توقف برنامه** در صورت بروز خطا  
✅ **مدیریت و رفع خطاها** بدون نیاز به خروج اضطراری  
✅ **امکان نمایش پیام سفارشی به کاربر** هنگام وقوع خطا  

# **۲. نحوه‌ی استفاده از `try-except`**  
🔹 **بخش `try`** → شامل کدی است که **ممکن است خطا ایجاد کند**.  
🔹 **بخش `except`** → خطا را دریافت می‌کند و **پاسخ مناسبی ارائه می‌دهد**.  

<div dir='ltr'>

```python
try:
    num = int(input("یک عدد وارد کنید: "))  # تبدیل ورودی به عدد صحیح
    print("عدد وارد شده:", num)
except ValueError:
    print("خطا! مقدار وارد شده عدد نیست.")
```

</div>

✅ اگر ورودی عدد باشد، مقدار چاپ می‌شود. 
✅ اگر ورودی غیر عددی باشد، except پیام مناسب نمایش می‌دهد.


# **۳. استفاده از `except` برای چند نوع خطا**  
🔹 می‌توان **چندین نوع خطای مختلف** را مدیریت کرد.  
🔹 هر `except` فقط **یک نوع خطا را پردازش می‌کند**.  

<div dir='ltr'>

```python
try:
    num = int(input("یک عدد وارد کنید: "))
    result = 10 / num  # تقسیم عدد ۱۰ بر مقدار ورودی
except ValueError:
    print("خطا! مقدار وارد شده عدد نیست.")
except ZeroDivisionError:
    print("خطا! تقسیم بر صفر مجاز نیست.")
```

</div>


# **۳. استفاده از `except` برای چند نوع خطا**  
🔹 می‌توان **چندین نوع خطای مختلف** را مدیریت کرد.  
🔹 هر `except` فقط **یک نوع خطا را پردازش می‌کند**.  

<div dir='ltr'>

```python
try:
    num = int(input("یک عدد وارد کنید: "))
    result = 10 / num  # تقسیم عدد ۱۰ بر مقدار ورودی
except ValueError:
    print("خطا! مقدار وارد شده عدد نیست.")
except ZeroDivisionError:
    print("خطا! تقسیم بر صفر مجاز نیست.")

```

</div>

✅ ValueError زمانی رخ می‌دهد که ورودی عددی نباشد. 
✅ ZeroDivisionError زمانی رخ می‌دهد که عدد ورودی 0 باشد.


# **۴. استفاده از `finally`**  
🔹 **بخش `finally` همیشه اجرا می‌شود**، چه خطا رخ دهد چه نه.  
🔹 مناسب برای **تمیزکاری و بستن منابع** مانند فایل‌ها و اتصالات پایگاه داده.  

<div dir='ltr'>

```python
try:
    file = open("data.txt", "r")  # باز کردن فایل
    content = file.read()
except FileNotFoundError:
    print("خطا! فایل یافت نشد.")
finally:
    file.close()  # بستن فایل

```

</div>

✅ بخش finally تضمین می‌کند که فایل بسته شود، حتی اگر خطا رخ دهد.


# **۵. استفاده از `else` همراه با `try-except`**  
🔹 **بخش `else` فقط در صورتی اجرا می‌شود که خطا رخ ندهد**.  
🔹 برای اجرای کد زمانی که **هیچ خطایی ثبت نشد**، مفید است.  

<div dir='ltr'>

```python
try:
    num = int(input("یک عدد وارد کنید: "))
except ValueError:
    print("خطا! مقدار وارد شده عدد نیست.")
else:
    print("عدد معتبر وارد شد:", num)
```
</div>

✅ اگر ورودی عددی باشد، else اجرا می‌شود. 
✅ اگر خطا رخ دهد، except اجرا شده و else نادیده گرفته می‌شود.


# **۶. نتیجه‌گیری**  
✅ **عبارت `try-except` ابزاری قدرتمند برای مدیریت خطاها و جلوگیری از توقف برنامه است.**  
✅ **می‌توان `except` را برای چندین نوع خطا استفاده کرد تا برنامه انعطاف‌پذیرتر شود.**  
✅ **بخش `finally` تضمین می‌کند که عملیات ضروری مانند بستن فایل‌ها انجام شود.**  
✅ **استفاده از `else` باعث مدیریت بهتر پردازش‌های بدون خطا می‌شود.**  

🚀 **با استفاده صحیح از `try-except`, می‌توان برنامه‌هایی قابل اعتماد و مقاوم در برابر خطاها ساخت!**  



</div>

````python
# برنامه‌ای برای ثبت و بررسی اطلاعات کاربران با استفاده از `try-except`

users = {}

def register_user():
    """ثبت نام کاربر و ذخیره اطلاعات در دیکشنری"""
    try:
        username = input("نام کاربری خود را وارد کنید: ")
        age = int(input("سن خود را وارد کنید: "))  # تبدیل به عدد صحیح
        users[username] = age  # ذخیره در دیکشنری
        print(f"کاربر {username} با سن {age} ثبت شد.")
    except ValueError:
        print("خطا! مقدار سن باید عدد باشد.")

def get_user_info():
    """دریافت اطلاعات کاربر ثبت‌شده"""
    try:
        username = input("نام کاربری موردنظر را وارد کنید: ")
        age = users[username]  # دریافت سن کاربر
        print(f"سن کاربر {username}: {age}")
    except KeyError:
        print("خطا! نام کاربری یافت نشد.")

# اجرای برنامه
while True:
    print("\n۱. ثبت نام کاربر\n۲. دریافت اطلاعات کاربر\n۳. خروج")
    choice = input("گزینه موردنظر را انتخاب کنید: ")
    
    if choice == "1":
        register_user()
    elif choice == "2":
        get_user_info()
    elif choice == "3":
        print("خروج از برنامه.")
        break
    else:
        print("گزینه نامعتبر! لطفاً دوباره امتحان کنید.")
````
