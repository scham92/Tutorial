# چرا پایتون برای هوش مصنوعی و علوم داده مناسب است؟  

پایتون یکی از محبوب‌ترین زبان‌های برنامه‌نویسی برای **هوش مصنوعی** و **علوم داده** است. برخی از دلایل کلیدی این محبوبیت عبارتند از:  

- **سادگی و خوانایی بالا**: پایتون با نحو ساده و قابل فهم خود، زمان توسعه را کاهش داده و تمرکز را بر روی حل مسائل علمی و داده‌محور نگه می‌دارد.  
- **کتابخانه‌های گسترده**: ابزارهایی مانند NumPy، Pandas، Matplotlib، Scikit-learn و TensorFlow قدرت پردازش داده و مدل‌سازی هوش مصنوعی را افزایش می‌دهند.  
- **جامعه فعال**: توسعه‌دهندگان و دانشمندان داده در سراسر جهان از پایتون استفاده کرده و به‌طور مستمر منابع، راهنماها و ابزارهای جدید ارائه می‌دهند.  
- **انعطاف‌پذیری بالا**: پایتون در طیف گسترده‌ای از کاربردها، از تحلیل داده و یادگیری ماشین تا پردازش تصویر و پردازش زبان طبیعی، قابل استفاده است.  

## رویکرد دوره

# نصب و مدیریت نسخه‌های Python  

## نصب Python در ویندوز  
برای نصب **Python** روی ویندوز، می‌توانید از لینک زیر نسخه مناسب را دریافت کنید:  
🔗 [دانلود Python برای ویندوز](https://www.python.org/downloads/windows/)  

بعد از دانلود، فایل اجرایی را باز کنید و هنگام نصب، گزینه **"Add Python to PATH"** را فعال کنید تا Python به مسیر سیستم اضافه شود.  

## نصب و بروزرسانی Python در لینوکس و macOS  
Python معمولاً به‌صورت **پیش‌فرض** روی لینوکس و macOS نصب است. برای بررسی نسخه نصب‌شده، از دستور زیر استفاده کنید:  

<div dir='ltr'>

```bash
python3 --version
```
</div> 

### بروزرسانی Python در لینوکس  
برای بروزرسانی Python، می‌توانید از **مدیر بسته** سیستم خود استفاده کنید:  

<div dir='ltr'>

- **Debian/Ubuntu:**  
  ```bash
  sudo apt update
  sudo apt install python3
  ```

- **Fedora:**  
  ```bash
  sudo dnf install python3
  ```
 

- **Arch Linux:**  
  ```bash
  sudo pacman -S python
  ```
</div>

# معرفی Jupyter Notebook  

<div align="center"> <img src="https://upload.wikimedia.org/wikipedia/commons/thumb/3/38/Jupyter_logo.svg/800px-Jupyter_logo.svg.png?20190118024747" alt="jupyter Logo" width="200"> </div>

**Jupyter Notebook** یک محیط تعاملی برای اجرای کد، تحلیل داده، و مستندسازی محاسبات است. این ابزار به کاربران اجازه می‌دهد تا **کد، متن توضیحی، فرمول‌های ریاضی، و نمودارها** را در یک سند واحد ترکیب کنند.  

## ویژگی‌های Jupyter Notebook  
- **پشتیبانی از چندین زبان برنامه‌نویسی** (مانند Python، R، و Julia)  
- **امکان اجرای کد به‌صورت تعاملی** و مشاهده خروجی در همان محیط  
- **قابلیت نمایش نمودارها و تصاویر** به‌صورت مستقیم در نوت‌بوک  
- **امکان مستندسازی و اشتراک‌گذاری تحلیل‌ها**  

## روش نصب سنتی Jupyter Notebook  

برای نصب **Jupyter Notebook** به روش سنتی، می‌توانید از **pip** استفاده کنید:  

<div dir='ltr'>

```bash
pip install notebook
```
</div>

پس از نصب، برای اجرای Jupyter Notebook دستور زیر را اجرا کنید:

<div dir='ltr'>

```bash
jupyter notebook
```
</div>

این دستور یک سرور محلی راه‌اندازی کرده و محیط Jupyter Notebook را در مرورگر باز می‌کند.

# افزونه Jupyter Notebook در Visual Studio Code

- **ویرایش و اجرای نوت‌بوک‌های Jupyter** بدون نیاز به اجرای جداگانه  
- **پشتیبانی از اجرای کد در سلول‌های جداگانه**  
- **امکان اتصال به سرورهای Jupyter راه دور**  

با این افزونه، می‌توانید از قدرت Jupyter بدون نیاز به ترک محیط VS Code بهره ببرید! 🚀

# معرفی بسته آناکوندا (Anaconda)  

<div align="center"> <img src="https://www.anaconda.com/wp-content/uploads/2022/12/anaconda_secondary_logo.svg" alt="Anaconda Logo" width="500"> </div>

**با نصب این بسته نیازی به نصب پایتون و کتابخانه های مرسوم علوم داده و هوش مصنوعی بصورت جداگانه نیست و همه به همراه ideهای محبوب در یک پکیج تجمیع شده اند.**

**Anaconda** یک توزیع محبوب از **Python** و **R** است که شامل ابزارهای قدرتمند برای **علوم داده، یادگیری ماشین، و تحلیل داده** می‌شود. این بسته شامل بیش از **300 کتابخانه** برای پردازش داده، یادگیری ماشین، و محاسبات علمی است.  

## ویژگی‌های Anaconda  
- **مدیریت آسان بسته‌ها و محیط‌های مجازی** با استفاده از `conda`  
- **پشتیبانی از ابزارهای محبوب** مانند Jupyter Notebook، Spyder، و TensorFlow  
- **سازگاری با سیستم‌عامل‌های مختلف** (ویندوز، لینوکس، macOS)  
- **امکان نصب و بروزرسانی کتابخانه‌ها** بدون نیاز به تنظیمات پیچیده  

## روش نصب Anaconda  

برای نصب **Anaconda**، می‌توانید نسخه مناسب را از لینک زیر دریافت کنید:  
🔗 **[دانلود Anaconda](https://www.anaconda.com/download/)**  

پس از دانلود و نصب، می‌توانید با دستور زیر محیط Anaconda را اجرا کنید:  

<div dir='ltr'>

```bash
conda --version
```
</div>

برای راه‌اندازی Jupyter Notebook در محیط Anaconda، از دستور زیر استفاده کنید:

<div dir='ltr'>

```bash
jupyter notebook
```
</div>

# معرفی Google Colab  

**Google Colab** (مخفف **Colaboratory**) یک محیط **Jupyter Notebook** مبتنی بر **ابر** است که توسط گوگل ارائه شده است. این ابزار به کاربران اجازه می‌دهد تا **کدهای پایتون** را بدون نیاز به نصب محلی اجرا کنند و از **پردازش‌های قدرتمند** مانند **GPU و TPU رایگان** بهره ببرند.  

🔗 **[دسترسی به Google Colab](https://colab.research.google.com/)**  


## مزایای Google Colab  
✅ **بدون نیاز به نصب** – نیازی به تنظیمات محلی یا نصب کتابخانه‌ها نیست.  
✅ **دسترسی رایگان به GPU و TPU** – امکان اجرای مدل‌های یادگیری ماشین با قدرت پردازشی بالا.  
✅ **ادغام با Google Drive** – ذخیره و بارگذاری نوت‌بوک‌ها به‌صورت مستقیم از گوگل درایو.  
✅ **اشتراک‌گذاری آسان** – امکان همکاری هم‌زمان روی نوت‌بوک‌ها مانند Google Docs.  
✅ **پشتیبانی از کتابخانه‌های محبوب** – شامل TensorFlow، PyTorch، NumPy، Pandas و موارد دیگر.  


## معایب Google Colab  
❌ **محدودیت منابع** – زمان اجرای محدود برای پردازش‌های سنگین (در نسخه رایگان).  
❌ **نیاز به اینترنت** – بدون اتصال اینترنت امکان استفاده وجود ندارد.  
❌ **محدودیت در ذخیره‌سازی داده‌ها** – داده‌های موقت پس از قطع اتصال از بین می‌روند.  
❌ **عدم کنترل کامل روی محیط** – برخی تنظیمات و نصب‌های خاص ممکن است محدود باشند.  



Google Colab یک ابزار عالی برای **یادگیری ماشین، علوم داده، و تحلیل داده‌ها** است، اما برای پروژه‌های سنگین ممکن است نیاز به نسخه **Colab Pro** یا راهکارهای دیگر داشته باشید. 🚀
