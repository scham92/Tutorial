# **۱. دستور `pip` در پایتون چیست؟**  
🔹 `pip` مخفف **"Python Package Installer"** است و برای **مدیریت و نصب کتابخانه‌های پایتون** استفاده می‌شود.  
🔹 به کمک `pip` می‌توان **کتابخانه‌های مورد نیاز را از مخزن `PyPI (Python Package Index)` دانلود و نصب کرد**.  
🔹 این ابزار به‌صورت **پیش‌فرض همراه با نسخه‌های جدید پایتون نصب شده است**.  

# **۲. نحوه‌ی نصب و استفاده از `pip`**  
✅ **بررسی نصب بودن `pip`**  


<div dir='ltr'>

```bash
pip --version
```

</div>


✅ نصب یک کتابخانه جدید


<div dir='ltr'>

```bash
pip install library_name
```

</div>


مثال:

<div dir='ltr'>

```bash
pip install numpy  # نصب کتابخانه NumPy
```

</div>


✅ بروزرسانی یک کتابخانه

<div dir='ltr'>

```bash
pip install --upgrade library_name
```


</div>

مثال:

<div dir='ltr'>

```bash
pip install --upgrade pandas  # بروزرسانی کتابخانه pandas
```

</div>


✅ حذف یک کتابخانه

<div dir='ltr'>

```bash
pip uninstall library_name
```

</div>



# **۳. کتابخانه‌های پایتون چیستند؟**  
🔹 **کتابخانه‌ها مجموعه‌ای از کدها و توابع آماده هستند** که قابلیت‌های جدیدی را به پایتون اضافه می‌کنند.  
🔹 برخی از کتابخانه‌های پرکاربرد پایتون:  
✅ **NumPy** → عملیات عددی و پردازش ماتریس‌ها  
✅ **Pandas** → تجزیه و تحلیل داده‌ها  
✅ **Matplotlib** → ترسیم نمودار  
✅ **Requests** → ارسال درخواست‌های HTTP  
✅ **TensorFlow** → هوش مصنوعی و یادگیری ماشین  
✅ **Django** → توسعه‌ی وب  

# **۴. نحوه‌ی `import` کردن کتابخانه‌ها در پایتون**  
🔹 برای استفاده از کتابخانه‌ها باید **آن‌ها را وارد (`import`) کنیم**.  
🔹 این کار با استفاده از دستور `import` انجام می‌شود.  

✅ **وارد کردن یک کتابخانه**  

<div dir='ltr'>

```python
import numpy  # وارد کردن کتابخانه NumPy
```

</div>


✅ وارد کردن یک کتابخانه با نام مستعار (alias)

<div dir='ltr'>

```python
import numpy as np  # استفاده از `np` به جای `numpy`
```

</div>


✅ وارد کردن یک تابع مشخص از یک کتابخانه

<div dir='ltr'>

```python
from math import sqrt  # فقط تابع `sqrt` را وارد می‌کند
print(sqrt(16))  # خروجی: 4.0
```

</div>


✅ وارد کردن چندین تابع از یک کتابخانه


<div dir='ltr'>

```python
from math import sqrt, pow
print(pow(2, 3))  # خروجی: 8.0
```


</div>


# **۵. نحوه‌ی `import` کردن توابع و کلاس‌های سایر فایل‌های پایتونی**  
🔹 می‌توان **توابع و کلاس‌های نوشته‌شده در سایر فایل‌های پایتونی را وارد کرد**.  
🔹 فایل‌ها باید در **یک مسیر (`directory`) باشند** یا مسیر آن‌ها مشخص شود.  

✅ **فرض کنیم فایل `utils.py` دارای تابع `greet()` است:**  

<div dir='ltr'>

```python
# فایل utils.py
def greet(name):
    return f"سلام {name}!"
```

</div>


✅ وارد کردن تابع از فایل دیگر:

<div dir='ltr'>

```python
from utils import greet  # وارد کردن تابع `greet`
print(greet("سجاد"))  # خروجی: سلام سجاد!
```

</div>


✅ وارد کردن کل فایل به‌عنوان یک ماژول

<div dir='ltr'>

```python
import utils
print(utils.greet("سجاد"))  # خروجی: سلام سجاد!
```

</div>


✅ مدیریت مسیر فایل‌های خارجی اگر فایل در مسیر متفاوتی باشد، می‌توان آن را به مسیر sys.path اضافه کرد:

<div dir='ltr'>

```python
import sys
sys.path.append('/مسیر/فایل')  # اضافه کردن مسیر دستی
import utils
```

</div>


# **۶. نتیجه‌گیری**  
✅ **دستور `pip` برای مدیریت کتابخانه‌های پایتون استفاده می‌شود.**  
✅ **کتابخانه‌ها قابلیت‌های جدیدی به پایتون اضافه می‌کنند.**  
✅ **با `import` می‌توان کتابخانه‌ها، توابع و کلاس‌ها را وارد کرد.**  
✅ **امکان `import` کردن فایل‌های دیگر پایتونی نیز وجود دارد.**  

🚀 **استفاده‌ی صحیح از `pip` و `import` باعث مدیریت بهتر پروژه‌ها و افزایش بهره‌وری در برنامه‌نویسی می‌شود!**  



````python
import numpy as np
import matplotlib.pyplot as plt

# تعریف ضرایب معادله درجه دو ax^2 + bx + c = 0
a = 1
b = -3
c = 2

# محاسبه ریشه‌ها با استفاده از فرمول درجه دو
roots = np.roots([a, b, c])
print(f"ریشه‌های معادله: {roots}")

# تولید نقاط x برای رسم نمودار
x = np.linspace(-2, 4, 100)
y = a*x**2 + b*x + c  # محاسبه مقادیر y برای هر x

# رسم نمودار
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f"${a}x^2 + {b}x + {c} = 0$", color="blue")
plt.axhline(0, color='black', linewidth=1)  # محور x
plt.axvline(roots[0], color='red', linestyle="--", label=f"first root ({roots[0]:.2f})")
plt.axvline(roots[1], color='green', linestyle="--", label=f"second root ({roots[1]:.2f})")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("حل معادله درجه دو و نمایش نمودار"[::-1])
plt.legend()
plt.grid(True)
plt.show()
````
