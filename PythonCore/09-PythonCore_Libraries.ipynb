# **€∂. ŸÜÿ™€åÿ¨Ÿá‚Äå⁄Ø€åÿ±€å**  
‚úÖ **ÿØÿ≥ÿ™Ÿàÿ± `pip` ÿ®ÿ±ÿß€å ŸÖÿØ€åÿ±€åÿ™ ⁄©ÿ™ÿßÿ®ÿÆÿßŸÜŸá‚ÄåŸáÿß€å Ÿæÿß€åÿ™ŸàŸÜ ÿßÿ≥ÿ™ŸÅÿßÿØŸá ŸÖ€å‚Äåÿ¥ŸàÿØ.**  
‚úÖ **⁄©ÿ™ÿßÿ®ÿÆÿßŸÜŸá‚ÄåŸáÿß ŸÇÿßÿ®ŸÑ€åÿ™‚ÄåŸáÿß€å ÿ¨ÿØ€åÿØ€å ÿ®Ÿá Ÿæÿß€åÿ™ŸàŸÜ ÿßÿ∂ÿßŸÅŸá ŸÖ€å‚Äå⁄©ŸÜŸÜÿØ.**  
‚úÖ **ÿ®ÿß `import` ŸÖ€å‚Äåÿ™ŸàÿßŸÜ ⁄©ÿ™ÿßÿ®ÿÆÿßŸÜŸá‚ÄåŸáÿßÿå ÿ™Ÿàÿßÿ®ÿπ Ÿà ⁄©ŸÑÿßÿ≥‚ÄåŸáÿß ÿ±ÿß Ÿàÿßÿ±ÿØ ⁄©ÿ±ÿØ.**  
‚úÖ **ÿßŸÖ⁄©ÿßŸÜ `import` ⁄©ÿ±ÿØŸÜ ŸÅÿß€åŸÑ‚ÄåŸáÿß€å ÿØ€å⁄Øÿ± Ÿæÿß€åÿ™ŸàŸÜ€å ŸÜ€åÿ≤ Ÿàÿ¨ŸàÿØ ÿØÿßÿ±ÿØ.**  

üöÄ **ÿßÿ≥ÿ™ŸÅÿßÿØŸá‚Äå€å ÿµÿ≠€åÿ≠ ÿßÿ≤ `pip` Ÿà `import` ÿ®ÿßÿπÿ´ ŸÖÿØ€åÿ±€åÿ™ ÿ®Ÿáÿ™ÿ± Ÿæÿ±Ÿà⁄òŸá‚ÄåŸáÿß Ÿà ÿßŸÅÿ≤ÿß€åÿ¥ ÿ®Ÿáÿ±Ÿá‚ÄåŸàÿ±€å ÿØÿ± ÿ®ÿ±ŸÜÿßŸÖŸá‚ÄåŸÜŸà€åÿ≥€å ŸÖ€å‚Äåÿ¥ŸàÿØ!**  



````python
import numpy as np
import matplotlib.pyplot as plt

# ÿ™ÿπÿ±€åŸÅ ÿ∂ÿ±ÿß€åÿ® ŸÖÿπÿßÿØŸÑŸá ÿØÿ±ÿ¨Ÿá ÿØŸà ax^2 + bx + c = 0
a = 1
b = -3
c = 2

# ŸÖÿ≠ÿßÿ≥ÿ®Ÿá ÿ±€åÿ¥Ÿá‚ÄåŸáÿß ÿ®ÿß ÿßÿ≥ÿ™ŸÅÿßÿØŸá ÿßÿ≤ ŸÅÿ±ŸÖŸàŸÑ ÿØÿ±ÿ¨Ÿá ÿØŸà
roots = np.roots([a, b, c])
print(f"ÿ±€åÿ¥Ÿá‚ÄåŸáÿß€å ŸÖÿπÿßÿØŸÑŸá: {roots}")

# ÿ™ŸàŸÑ€åÿØ ŸÜŸÇÿßÿ∑ x ÿ®ÿ±ÿß€å ÿ±ÿ≥ŸÖ ŸÜŸÖŸàÿØÿßÿ±
x = np.linspace(-2, 4, 100)
y = a*x**2 + b*x + c  # ŸÖÿ≠ÿßÿ≥ÿ®Ÿá ŸÖŸÇÿßÿØ€åÿ± y ÿ®ÿ±ÿß€å Ÿáÿ± x

# ÿ±ÿ≥ŸÖ ŸÜŸÖŸàÿØÿßÿ±
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=f"${a}x^2 + {b}x + {c} = 0$", color="blue")
plt.axhline(0, color='black', linewidth=1)  # ŸÖÿ≠Ÿàÿ± x
plt.axvline(roots[0], color='red', linestyle="--", label=f"first root ({roots[0]:.2f})")
plt.axvline(roots[1], color='green', linestyle="--", label=f"second root ({roots[1]:.2f})")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("ÿ≠ŸÑ ŸÖÿπÿßÿØŸÑŸá ÿØÿ±ÿ¨Ÿá ÿØŸà Ÿà ŸÜŸÖÿß€åÿ¥ ŸÜŸÖŸàÿØÿßÿ±"[::-1])
plt.legend()
plt.grid(True)
plt.show()
````
